```python
def findStable(X,Y,countGap = 5,gap=0.3,errorGap = 5,mean_range=10,p_flag = False):
    # Assume far and end come up one by one,start from far
    circleX=[]
    circleY=[]
    count = 0
    start = 0
    stable = []
    stable_close=[]
    stable_far=[]
    errorCount = 0
    close_flag = True
    
    # Find Stable: Algorithm
    for i in range(1,len(Y)):
        if abs(Y[i]-Y[i-1])<gap and Y[i]>10:
            count = count+1
            errorCount = 0
            if start == 0:
                start = i
        elif abs(Y[i]-Y[i-1])>gap:
            errorCount = errorCount + 1
            if errorCount <= errorGap:
                continue
            if count>=countGap:
                end = i - errorCount - 1
                stable.append((start,end))
                center = end-int(countGap/2)
                circleX.append(X[center])
                circleY.append(Y[center])
                if close_flag:
                    stable_close.append((start,end))
                    close_flag = False
                else:
                    stable_far.append((start,end))
                    close_flag = True
            count = 0
            start = 0
            
    # Calculate Average
    result = calculateAvg(X,Y,stable,mean_range=mean_range,p_flag = p_flag)
    result_close = calculateAvg(X,Y,stable_close,mean_range=mean_range)
    result_far = calculateAvg(X,Y,stable_far,mean_range=mean_range)
    
    return circleX,circleY,result,result_close,result_far,stable_close,stable_far
```

```python
def calculateAvg(X,Y,stable,mean_range=10,p_flag = False):
    npY = np.array(Y)
    result = np.array([])
    for s in stable:
        Y_target = Y[s[1]-mean_range+1:s[1]+1]
        mean = np.mean(Y_target)
        result=np.append(result,mean)
        if p_flag:
            print("Time Interval(s):%.4f-%.4f"%(X[s[0]],X[s[1]]))
            print("Last "+str(mean_range)+" point average(mV) = "+str(mean))
            plotDifference(X,Y,s)
    return result
```

```python
def plotStable(abf,X,Y,circleX,circleY,size):
    plt.figure(figsize = (18,5))
    plt.plot(X,Y)
    plt.scatter(circleX,circleY,s=(size/4)**2,facecolors='none', edgecolors='r')
    plt.ylabel(abf.sweepLabelY)
    plt.xlabel(abf.sweepLabelX)
    plt.show()
```

```python
def plotDifference(X,Y,s):
    start = s[0]
    end = s[1]
    x = X[start:end]
    y = Y[start:end]
    dif = []
    for i in range(1,len(y)):
        differ = abs(y[i]-y[i-1])
        dif.append(differ)
    plt.figure()
    plt.hist(dif,bins=100)
```

```python
def findStable_byM(X,Y,countGap = 5,gap=0.3,errorGap = 5,mean_range=10,p_flag = False):
    # Assume far and end come up one by one,start from far
    m = []
    extreamX = []
    extreamY = []
    # Plot m
    for i in range(1,len(Y)):
        m.append((Y[i]-Y[i-1])/(X[i]-X[i-1]))
        if (Y[i]-Y[i-1])/(X[i]-X[i-1]) == 0:
            extreamX.append(X[i-1])
            extreamY.append(Y[i-1])
    plt.figure(figsize=(18,5))          
    plt.plot(X[:-1],m)
    plt.scatter(extreamX,extreamY,s = 5**2,facecolors='none', edgecolors='r')
    plt.show()
    
    plt.figure(figsize=(18,5))
    plt.plot(X[:-1],Y[:-1])
    plt.show()
    
    plt.figure()
    plt.hist(m,bins = 200)
    plt.show()
```

```python
timeStart = 180000
timeEnd = 190000
X = abf.sweepX[timeStart:timeEnd]
Y = abf.sweepY[timeStart:timeEnd]
countGap = 50
gap = 0.7
errorGap = 5
mean_range = 10
p_flag = False

cirX,cirY,result,far,end,_,_ = findStable(X,Y,countGap = countGap,gap = gap,errorGap = errorGap,mean_range = mean_range,p_flag = p_flag)
plotStable(abf,X,Y,cirX,cirY,countGap)
```

```python
# Separate AUT and Quartz
# Total 25 rows, first row contains 4 + 100 * 6 = 604
# 2-25 contains (5+100)*24 = 2520
def separate(arr,pqty = 40):
    # Ignore first row
    arr = arr[604:]
    arr = np.reshape(arr,(24,105))
    aut = np.array([])
    quartz = np.array([])
    for row in arr:
        aut = np.append(aut,row[5:5+pqty])
        quartz = np.append(quartz,row[-pqty:])
    return aut,quartz
```

```python
bins = 50
aut_close,quartz_close = separate(close)
aut_far,quartz_far = separate(far)


plt.figure()
plt.hist((aut_close,quartz_close,aut_far,quartz_far),bins=bins,label = ('EMF close-AUT','EMF close-QUARTZ','EMF far-AUT','EMF far-QUARTZ'),color = ('red','blue','darkorange','green'))
plt.legend()
plt.title("50-value mean histogram")
plt.show()

plt.figure()
plt.hist((aut_close-aut_far,quartz_close-quartz_far),bins=bins,label = ('AUT','QUARTZ'),color = ('red','blue'))
plt.legend()
plt.title("50-value mean,close-far histogram")
plt.show()

plt.figure()
plt.hist((aut_close/aut_far,quartz_close/quartz_far),bins=bins,label = ('AUT','QUARTZ'),color = ('red','blue'))
plt.legend()
plt.title("50-value mean,close/far histogram")
plt.show()
```

```python
def plot_colormap(data):
    """
    Helper function to plot data with associated colormap.
    """
   # fig, axs = plt.subplots(1, n, figsize=(n * 2 + 2, 3),
    #                        constrained_layout=True, squeeze=False)
    figure, axes = plt.subplots(figsize=(data.shape[1]/5,data.shape[0]/5))
    psm = axes.pcolormesh(data, cmap='rainbow',rasterized=True)
    figure.colorbar(psm, ax=axes)
    axes.invert_yaxis()
    plt.show()
```
